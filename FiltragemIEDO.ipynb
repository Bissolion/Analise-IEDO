```python
import soccerdata as sd
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
csvnome = "ESP-La Liga_2023_PLAYERS.csv"
df = pd.read_csv(csvnome)
```

```python
print(list(df.columns))
```

```python
df = df[df['Playing Time_Min'] >= 900]
df['Eficiência'] = df["Tackles_TklW"] / df["Tackles_Tkl"]
df['Eficiência'] = df['Eficiência'].fillna(0)

df["IEDO"] = (df["SCA Types_Def"] + df["SCA Types_Def"] + df["Tackles_Att 3rd"]) * df['Eficiência']
df['Eficiência']
# Ordenar pela nova métrica em ordem decrescente
df = df.sort_values(by='IEDO', ascending=False)
df.index = range(1, len(df) + 1)
df.head(10)
```

```python
df_filtro = df[["player_","pos_","team_","season_","Tackles_Tkl","Tackles_TklW","Tackles_Att 3rd","SCA Types_Def", "GCA Types_Def","Eficiência", "IEDO"]] 
df_filtro.head(10)
```

```python
df_filtro["Eficiência"] = df_filtro.get("Eficiência", 0).round(2)
df_filtro["IEDO"] = df_filtro.get("IEDO", 0).round(2)
df_filtro.head()
```

```python
df_filtro.rename(columns={
    "player_": "Jogador",
    "pos_": "POS",
    "team_": "Time",
    "season_": "Temporada",
    "Tackles_Tkl": "Desarmes",
    "Tackles_TklW": "Desarmes Ganhos",
    "Tackles_Att 3rd": "Desarmes no Terço Final",
    "SCA Types_Def": "SCA DEF",
    "GCA Types_Def": "GCA DEF"
}, inplace=True)
df_filtro.to_csv("FILTROIEDO_"+ csvnome)
df_filtro.head(10)
```
