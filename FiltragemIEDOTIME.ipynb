```python
import soccerdata as sd
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
csvnome = "ENG-Premier League_2024_TEAMS.csv"
df = pd.read_csv(csvnome)
```

```python
print(list(df.columns))
```

```python
df = df[df['Playing Time_Min'] >= 900]
df['EficiÃªncia'] = df["Tackles_TklW"] / df["Tackles_Tkl"]
df['EficiÃªncia'] = df['EficiÃªncia'].fillna(0)

df["IEDO"] = (df["SCA Types_Def"] + df["SCA Types_Def"] + df["Tackles_Att 3rd"]) * df['EficiÃªncia']
df['EficiÃªncia']
# Ordenar pela nova mÃ©trica em ordem decrescente
df = df.sort_values(by='IEDO', ascending=False)
df.index = range(1, len(df) + 1)
df.head(10)
```

```python
df_filtro = df[["team_","season_","Tackles_Tkl","Tackles_TklW","Tackles_Att 3rd","SCA Types_Def", "GCA Types_Def","EficiÃªncia", "IEDO"]] 
df_filtro.head(10)
```

```python
df_filtro["EficiÃªncia"] = df_filtro.get("EficiÃªncia", 0).round(2)
df_filtro["IEDO"] = df_filtro.get("IEDO", 0).round(2)
df_filtro.head()
```

```python
df_filtro.rename(columns={
    "team_": "Time",
    "season_": "Temporada",
    "Tackles_Tkl": "Desarmes",
    "Tackles_TklW": "Desarmes Ganhos",
    "Tackles_Att 3rd": "Desarmes no TerÃ§o Final",
    "SCA Types_Def": "SCA DEF",
    "GCA Types_Def": "GCA DEF"
}, inplace=True)
df_filtro.to_csv("FILTROIEDO_"+ csvnome)
df_filtro.head(10)
```
